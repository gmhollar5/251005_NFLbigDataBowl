# Import Libraries

```python
import pandas as pd
import os
import glob

pd.set_option('display.max_columns', None)  # Show all columns
```

# Read in Data

```python
train_data_path = "/kaggle/input/nfl-big-data-bowl-2026-prediction/train"
```

```python
# Input files
input_files = os.path.join(train_data_path, 'input*.csv')

# List to hold each DataFrame
all_dfs = []

# Loop through matching files
for file_path in glob.glob(input_files):
    print(f"Reading: {file_path}")
    df = pd.read_csv(file_path)
    all_dfs.append(df)

# examine files read
print(f"{len(all_dfs)} files read")
    
# Combine all into one DataFrame
combined_df = pd.concat(all_dfs, ignore_index=True)

# Optional: Show info or preview
print("Combined DataFrame shape:", combined_df.shape)
print(combined_df.head())
```
